## **Presentación del ejercicio.**
La función a la cual se le debe aproximar el área bajo la curva es:

$$f(x) = {e} ^ {(\frac{x}{2})} + 1 $$

En el intervalo $[0,3]$ con $n = 4, 10, 785, 2785$ particiones.

#<center> <h1><B>Desarrollo</h1> </center>




#**Paso 1: funcion pera desarrollar sumas de Riemman(Izquierdas).**

La siguiente solucion se plantea a partir del uso de las sumas de Riemann por la izquierda donde A es el area bajo la curva de la función a trabajar.

$$A = \displaystyle\sum_{i=1}^{n} f(x_i)\Delta{x}$$

##**Paso 1: Creación de la función para desarrollar sumas de Riemann Izquierdas.**

```python
import math
def sumasRiemann(a,b,n):

  sumarR = 0
  #Generar el delta el cual es la base de los rectangulos.
  delta = (b-a)/n
  for i in range(0,n,1):
    #Paso 1 eneración de la particion.
    x = a + (i*delta)
    #paso 2: Ingreso de la funcion y evaluacion de los elementos de la partición.
    y = pow(math.e,(x/2))+1
    #paso 3: Multiplicar delta o base por alturas y asi encontramos el area del rectangulo
    suma = y * delta
    #paso 4 sumar el area de los rectangulos.
    sumarR = sumarR + suma
  #Paso 5 realizamos en retorno del area.
  return sumarR

# Llamado a la función cuando para 4 particiones.
print("La suma de Riemman es: ", sumasRiemann(0,3,4))

# Llamado a la función cuando para 10 particiones.
print("La suma de Riemman es: ", sumasRiemann(0,3,10))

# Llamado a la función para 3 ultimos numeros de la cedula particiones.
print("La suma de Riemman es: ", sumasRiemann(0,3,785))
```

###**Paso 2: Crear la funcion que me imprime las sumas de Riemman llamando a la función sumasRiemann.**


```python
def sumaDeRiemann():
  a = float(input("Ingrese el limite inferior del intervalo:\n"))
  b = float(input("Ingrese el limite superior del intervalo:\n"))
  n = int(input("Ingrese el numero de particiones n = :\n"))

  print("La suma de Riemman es: ", sumasRiemann(a,b,n))

  return sumasRiemann(a,b,n)

sumaDeRiemann()
```

##**Paso 3: Resultado de las 4 aproximaciones solicitadas**

```python
SDR4 = []

for i in range(0,4,1):
  x = sumaDeRiemann()
  SDR4.append(x)

print("las cuatro aproximaciones son: ",SDR4)
```

##**Paso 4: Generar los 4 errores.**

#*Error(%) = $\frac{|Valor\ exacto  \ - Valor\ aproximado|}{Valor\ exacto}$ x 100%*


```python
error4 = []
for i in range(0,4,1):
  x = (abs(9.96 - SDR4[i])/9.96)*100
  error4.append(x)

print("Los cuatro errores son ",error4)
```

#**Paso 5: Generar la tabla.**

```python
from tabulate import tabulate

SumasRiemann = [["4",SDR4[0],error4[0]],["10",SDR4[1],error4[1]],["785",SDR4[2],error4[2]],["2785",SDR4[3],error4[3]]]
print(tabulate(SumasRiemann, headers=["Partición","Sumas","Error(%)"], tablefmt='francy_grid'))
```
